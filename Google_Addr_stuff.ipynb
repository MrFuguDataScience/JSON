# Obtaining Data:

+ The data came from `Google` and was sent to me raw. I converted to a json file for use and needed to utilize `json_normalize` to flatten out these data for interpretation and investigation.

https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_json.html

```python
# with open('google_addr_file.json', 'w') as f_out:
#     json.dump(data , f_out)

googs_=pd.read_json('google_addr_file.json')


# df_=pd.DataFrame(googs_['timelineObjects'])

ext_df_=pd.json_normalize(json.loads(googs_['timelineObjects'].to_json(orient="records")))

ext_df_
```

# The Desired Variables of Interest:

+ These consist of nested json objects separated by dot notation. 
    + Each dot is a new layer deeper into the nesting.

```python
desired_vars=['activitySegment.duration.startTimestampMs',
       'activitySegment.duration.endTimestampMs', 
       'activitySegment.distance',
       'activitySegment.activities',
       'placeVisit.location.address',
       'placeVisit.duration.startTimestampMs',
       'placeVisit.duration.endTimestampMs',
       'activitySegment.endLocation.address',
       'activitySegment.endLocation.locationConfidence',
        'placeVisit.location.name',
        'activitySegment.endLocation.name']
```

```python
# Parse Columns of Interest:
df_start=ext_df_.loc[:,desired_vars]
df_start.dtypes
```

# Explode Column: 
+ This will take the column of interest and unnest the dictionaries. As an artifact we
need to establish that this will increase the number of rows. This occurs because the nested 
values will each get a row. 

`Note: this is useful if you need this format or want long data`

```python
df_start_=pd.json_normalize(json.loads(df_start.explode('activitySegment.activities').to_json(orient="records")))
df_start_.head()

'''
 These activitySegment.activities.activityType & activitySegment.activities.probability
 are expanded out. and there are 1 or 2 residual columns with all Nans created
 that you should remove. This is because they are the outer layers that don't 
 store anything.
''' 
```

# The Timestamps: will need to be converted from `Ms` to an actual useable format for humans

+ There are four columns of interest: 

`activitySegment.duration.startTimestampMs,	activitySegment.duration.endTimestampMs,placeVisit.duration.startTimestampMs,	placeVisit.duration.endTimestampMs`

**As a disclaimer**, I chose to find the strings containing `Timestamp` instead of directly calling these columns for two reasons:
+ If you have new data or more columns than your able to directly view it would be beneficial to not hard code your results.
+ Also, it gives you an option to work with and builds new skills 

```python

ste=[] # store the columns related to timestamp in their string names
for i in df_start:
    if i.find('Timestamp')>0:
        ste.append(i)
#     else:
#         ste.append(0) # use this if you want to preserve order of list
len(ste)

df_fill=df_start.loc[:,ste].fillna(0) # fill Na with zero

df_fill=df_fill.apply(pd.to_numeric) # change datatype

# from collections import defaultdict

tmst=defaultdict(list)

for i in df_fill.to_dict().items():
    for j in i[1].values():
        if j == 0:
            tmst[i[0]].append('Nada') # if timestamp absent replace with 'nada'
        else:
            tmst[i[0]].append(datetime.fromtimestamp(j/1000)) # divide by 10^-3 sec

# formatted: Timestamps, can be further processed depending on your needs            
df_start[ste]=pd.DataFrame(tmst)

df_start
```

# Float values for the `Confidence and activity distance` need to be changed to zero where Nan exist

```python
df_start[['activitySegment.endLocation.locationConfidence',
'activitySegment.distance']]=df_start.loc[:,['activitySegment.endLocation.locationConfidence',
'activitySegment.distance']].fillna(0)

df_start
```

```python
# convert the new column by filling Na with 'Nope'
df_start['activitySegment.activities']=df_start['activitySegment.activities'].fillna('Nope')
```

# `Need to Extract the data from list of dictionaries in [activitySegment.activities] column`

```python
new_df=pd.json_normalize(json.loads(df_start.explode('activitySegment.activities').to_json(orient="records")))

# len(set(new_df['activitySegment.activities.activityType']) )    

tt=[]
# showing datatypes so I can figure out how to deal with these data
for i in df_start['activitySegment.activities'].values.tolist():
    print(type(i))
```

```python

# creating a representation of original data to remove rows of NaN, to parse easily
trick_=[{'activityType': 'Nada', 'probability': 'Nada'}]*len(df_start['activitySegment.activities'][0])

hack_=[] # used to convert Nan to correct format for ease of use
for i in df_start['activitySegment.activities'].values.tolist():
    if type(i)==str: 
        hack_.append(trick_)
    else:
        hack_.append(i)

# new DF without rows of Nan or strings and all same format: dict lists
no_nan_LstofDict=pd.DataFrame(hack_)

# flatten columns of dictionaries to new columns:
activity_n_prob=pd.json_normalize(json.loads(no_nan_LstofDict.to_json(orient="records")))

'''
Find any string with this word, >0 since we are dealing with location of this word
if i doesn't appear you get -1

'''


act=[]
prob=[]
for i in activity_n_prob:
    if i.find('activit')>0: 
        act.append(i)
    else:
        prob.append(i)

Prob_=[]       

for i in activity_n_prob[prob].values:
    Prob_.append(i)


Act_=[]
for j in activity_n_prob[act].values:
    Act_.append(j)


activityKey_Probval=defaultdict(list) # key, with values as a list


u=[] # list of probabilities
for i in Prob_:
    u.extend(i)


v=[] # list of activities
for i in [Act_[0]]*len(df_start): # copying the keys so we have enough to do dict()
    v.extend(i)
# len(v)

for i in list(zip(v,u)):
    activityKey_Probval[i[0]].append(i[1])

```

```python
# activities column label and value is the probability of it occuring 
activities_df=pd.DataFrame(activityKey_Probval)
activities_df
```

# Fix Column Names: 
+ Remove dot notation and take last occurance

```python
col_names=[] # store the new column names
for i in df_start.columns:
    col_names.append(i.split('.',1)) # split on first occurance of a dot
col_names


fin_col_names=[sublist[-1] for sublist in col_names] # take last entry of each list

df_start.columns=fin_col_names # change to new names
df_start
```

```python
# Final DF: put together column wise
pd.concat([df_start,activities_df],axis=1)
```
